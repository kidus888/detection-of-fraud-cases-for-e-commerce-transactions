```python
import sys
sys.path.append('../scripts')
```

```python
import pandas as pd
from data_preprocessing import handle_missing_values, remove_duplicates, correct_data_types, map_ip_to_country, create_transaction_frequency, create_time_features, normalize_and_scale
```

```python
# Load datasets
fraud_data = pd.read_csv('../data/Fraud_Data.csv')
ip_data = pd.read_csv('../data/IpAddress_to_Country.csv')
```

```python
# Data Cleaning
fraud_data = handle_missing_values(fraud_data)
fraud_data = remove_duplicates(fraud_data)
fraud_data = correct_data_types(fraud_data)
```

```python
# IP Address to Country Mapping
fraud_data = map_ip_to_country(fraud_data, ip_data)
```

```python
# Feature Engineering
fraud_data = create_transaction_frequency(fraud_data)
fraud_data = create_time_features(fraud_data)
```

```python
# Normalization and Scaling
fraud_data = normalize_and_scale(fraud_data)
```

```python
# Exploratory Data Analysis (EDA)
print(fraud_data.describe())
print(fraud_data.info())
```
